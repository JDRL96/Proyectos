```julia
using SymPy
```

```julia
G, M, t, r, θ, ϕ, J  =  symbols("G, M, t, r, theta, phi, J")
```

```julia
a = J/M
```

```julia
ρ = sqrt(r^2 + (a*cos(θ))^2)
```

```julia
Δ = r^2 - 2*G*M*r +a^2
```

```julia
g = [-(1-(2G*M*r)/(ρ^2)) 0 0 -2*G*M*a*r*((sin(θ)/ρ)^2)
     0 (ρ^2)/Δ 0 0
     0 0 ρ^2 0
     -2*G*M*a*r*((sin(θ)/ρ)^2) 0 0 (sin(θ)/ρ)^2*((r^2 + a^2)^2 - a^2*Δ*(sin(θ))^2)]
```

```julia
simplify.(g)
```

```julia
"Variables"
X = [t, r, θ, ϕ]
```

```julia
X[1]
```

```julia
g_inv = simplify.(inv(g))
```

```julia
"Símbolos de Christoffel"
Γ=[]
for i in 1:4
    push!(Γ, [sum([simplify(g_inv[i,m]*(diff(g[m,k],X[l]) + diff(g[m,l],X[k])- diff(g[k,l],X[m]))) for m in 1:4]) for k in 1:4, l in 1:4]/2)   
end    
```
